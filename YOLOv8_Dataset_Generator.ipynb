```python
import random
from PIL import Image, ImageEnhance
import os
```

```python
folder_Path_Base = "./FruitsTest/Base/"
folder_Path_Object = "./FruitsTest/Object/"


image_files_Base = [os.path.join(folder_Path_Base, f) for f in os.listdir(folder_Path_Base) if f.endswith(('.png', '.jpg', '.jpeg', '.gif'))]
image_files_Object = [os.path.join(folder_Path_Object, f) for f in os.listdir(folder_Path_Object) if f.endswith(('.png', '.jpg', '.jpeg', '.gif'))]
```

```python
i = 1
img_count = 10000
object_image_size = range(100, 300, 5)


# Generate Trainning Image
while( i < img_count ):
    # Bacground Image Select
    base_image = Image.open(random.choice(image_files_Base))
    base_image = base_image.resize([640,640]) # input of YOLOv8
    base_image_editable = base_image.copy()

    # Object Pices count
    object_count = random.randint(1, 10)

    # Label file
    f = open("./Datasets/Train/labels/" + str(i) + ".txt", "w")

    # Open overlayimage
    for j in range(object_count) :
        # Select Object Image
        file_name = random.choice( image_files_Object ) 
        object_image = Image.open( file_name )
        object_image = object_image.convert("RGBA")

        # Choice Object imge size & Resize
        object_size = random.choice(object_image_size)
        object_image = object_image.resize([object_size, object_size])

        # Set Image Bound
        max_bound = 640 - object_size

        # Set Image Location
        # Left-Upper Corner
        rand_xpos = random.randint(0, max_bound)
        rand_ypos = random.randint(0, max_bound)

        # Convert Image ( Rotate, Lightness, Sharpness )
        object_image = object_image.rotate(random.randint(0,360), expand = 1)

        object_image_enhancer = ImageEnhance.Contrast(object_image)
        object_image = object_image_enhancer.enhance(random.uniform(0.5,1.5))

        object_image_enhancer = ImageEnhance.Color(object_image)
        object_image = object_image_enhancer.enhance(random.uniform(0.5,1.5))

        object_image_enhancer = ImageEnhance.Sharpness(object_image)
        object_image = object_image_enhancer.enhance(random.uniform(0.5,1.5))


        # Paste Image
        alpha = object_image.split()[3]
        Image.Image.paste( base_image_editable , object_image, (rand_xpos, rand_ypos), mask = alpha)
        

        # Add Labels
        file_name = file_name.split('/')
        file_name = file_name[len(file_name)-1]
        file_name = list(file_name)
        object_type = file_name[0]

        
        # Position Ratio
        x_pos = float((rand_xpos + object_size/2)) / 640
        y_pos = float((rand_ypos + object_size/2)) / 640
        size_ratio = float(object_size)/ 640

        label = object_type + " " + str(x_pos) + " " +  str(y_pos) + " " +  str(size_ratio) + " " +  str(size_ratio)
        f.write(label + "\n")



    
    # Save the result
    base_image_editable.save("./Datasets/Train/imges/" + str(i) + ".jpg", 'JPEG')
    
    f.close()
    
    i = i + 1
```
